```python
# plot out a singel frame
frame = np.load('/Users/macbook/GH/cira/data/IRTL_frames/IRTL_frame_0400.npy')
plt.figure(figsize = [10,10])
plt.imshow(frame)

## Now choose several specific nodes in the frame to test our result:
node = np.array([[40,700] , [200,600], [300,125], [600,50], [800,330], [800,550]])
plt.scatter(node.T[0], node.T[1], color = 'w', linewidths=25)
```

```python
def pointer(node):
    '''temperature file is of shape: [708, 720, 968]
    with reshpe, we transfered that to [708, 720*968]
    while 720 refers to the length of y axis of the frame and 968 refers to that of x axis
    note that pixel location starts from 0
    '''
    return node[1]*720 + node[0]

node_reshape_pos = map(pointer, node)
```

```python
# now we plot out result data of selected pixel by time
# node0

i = 0
plt.figure(figsize=[10,10])

plt.subplot(321)
plt.plot(res[1][node_reshape_pos[i] ,:], label = 't = 1')
plt.legend()

plt.subplot(322)
plt.plot(res[2][node_reshape_pos[i] ,:], label = 't = 2')
plt.legend()

plt.subplot(323)
plt.plot(res[3][node_reshape_pos[i] ,:], label = 't = 3')
plt.legend()

plt.subplot(324)
plt.plot(res[4][node_reshape_pos[i] ,:], label = 't = 4')
plt.legend()

plt.subplot(325)
plt.plot(res[5][node_reshape_pos[i] ,:], label = 't = 5')
plt.legend()

plt.subplot(326)
plt.imshow(frame)
plt.scatter(node.T[0][i], node.T[1][i], color = 'w', linewidths=30)
```
