```python
# if we run clustering directly on single pixels, it would cause memory error
# try take average over each 10 pixels
res5_ave = res5.reshape([10, 69696,708]).mean(axis = 0)
```

```python
km = KMeans(n_clusters=5)
km.fit(res5_ave[:10000])
km_labels = km.labels_

res5_ave = pd.DataFrame(res5_ave[:10000])
res5_ave['kmlabel'] = km_labels
```

```python
print 'number of pixel in each cluster'
res5_ave.groupby('kmlabel').count()[0]
```

```python
gp_res5_ave = res5_ave.groupby('kmlabel')
```
